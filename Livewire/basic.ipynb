# Why livewire 

     Livewire is used in Laravel projects to build dynamic, reactive, and interactive user interfaces without writing significant JavaScript. It allows developers to 
create modern, single-page application (SPA)-like experiences while keeping the simplicity of server-side rendering. 
 [ No need of Frontend Frameworks like vue.js or React JS]


 # Install livewire

    > composer require livewire/livewire

 - Livewire works everything is done using components .. So for the example search bar is component , if you want a table its going to be a seperate component, so everything done through components.
